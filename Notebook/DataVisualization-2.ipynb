```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd

df = pd.read_csv(r'/Users/youssefragai/Desktop/DS/ipCountAll.csv')
df.head()
```

```python
#IP Count All
IP = df['IP']
Freq = df['Freq']
 
# Figure Size
fig = plt.figure(figsize=(25, 20))
 
# Horizontal Bar Plot
plt.bar(IP, Freq, ec='blue')
 
    
plt.xlabel("IP")
plt.ylabel("Frequency")
plt.title("IP traffic")
# Show Plot
plt.show()
```
